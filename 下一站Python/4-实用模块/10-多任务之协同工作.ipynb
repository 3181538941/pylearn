# 多任务之协同工作
## 一个缩略图拼图工作
- 文件夹中有1400个JPG文件；
- 逐个制作为(256,192)的缩略图，并将文件日期标注在缩略图中；
- 再将所有缩略图拼到一个大图里，保存为JPG文件输出

![](mktilemap.jpg)

## 与批量缩略图制作（蚂蚁搬家）对比
- 批量缩略图制作是“各自为战”，任务没有交集，大图需要拼接为一体；
- 批量缩略图制作输出大量文件，大图仅需要1个文件；
- 批量缩略图制作中每个缩略图完成即可结束当前工作，大图需要全部完成才结束。

## 工作流程
- 引入“共享队列”，多个进程可以并发存取队列；
- 进程分为一个分派工作的“经理”和多个分头工作的“工人”；
- “经理”将工作指令放进“任务队列”；
- “工人”从“任务队列”取到工作指令，完成工作后将结果放入“成果队列”；
- “经理”从“成果队列”取到每个缩略图，拼接入大图；
- 等所有“工人”完成工作，“经理”宣告任务完成
