# 准备测试数据

````python
from random import randint, uniform, shuffle
import matplotlib.pyplot as plt
plt.style.use("seaborn")
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
%matplotlib inline

nmin, nmax, total = 1, 100, 50
nlist = [randint(nmin, nmax) for n in range(total)]  # 一系列随机整数
flist = [uniform(nmin, nmax) for n in range(total)]  # 一系列随机浮点数
slist = "Alfa, Bravo, Charlie, Delta, Echo, Foxtrot, Golf, Hotel, India, "\
    "Juliett, Kilo, Lima, Mike, November, Oscar, Papa, Quebec, Romeo, Sierra, "\
    "Tango, Uniform, Victor, Whiskey, X-ray, Yankee, Zulu".split(", ")
shuffle(slist)  # 对字符串列表进行随机"洗牌"
print("#整数：", nlist)
print("#浮点数：", ", ".join([f"{f:.2f}" for f in flist]))
print("#字符串：", slist)
````

````python
plt.figure(figsize=(8,3), dpi=150)
plt.subplot(1, 2, 1)
plt.title("nlist整数")
plt.bar(range(total), nlist)
plt.subplot(1, 2, 2)
plt.title("flist浮点数")
plt.stem(range(total), flist)
plt.show()
````

# 用内置函数```sorted```排序
```python
sorted(iterable, /, *, key=None, reverse=False)
```
## 整数排序

````python
nsorted = sorted(nlist)  # 对整数列表进行排序

print("整数排序", type(nsorted))
print(nsorted)
````

````python
plt.figure(figsize=(8,3), dpi=150)
plt.subplot(1, 2, 1)
plt.bar(range(total), nlist)
plt.subplot(1, 2, 2)
plt.bar(range(total), nsorted)
plt.show()
````

## 浮点数排序

````python
fsorted = sorted(flist)  # 对浮点数列表进行排序

print("浮点数排序", type(fsorted))
print(", ".join([f"{f:.2f}" for f in fsorted]))
````

````python
plt.figure(figsize=(8,3), dpi=150)
plt.subplot(1, 2, 1)
plt.stem(range(total), flist)
plt.subplot(1, 2, 2)
plt.stem(range(total), fsorted)
plt.show()
````

## 字符串排序

````python
ssorted = sorted(slist)  # 对字符串列表进行排序

print("字符串排序", type(ssorted))
print(ssorted)
````

# 在原列表中就地排序
- 列表类型的方法```list.sort```
- **不**生成新列表对象，在原列表中就地排序

````python
nlist_copy = nlist[:]  # 为了不改变nlist，我们从nlist做一个拷贝到nlist_copy
print("原整数列表", id(nlist_copy))
print(nlist_copy)

nlist_copy.sort()  # 在nlist_copy中就地排序

print("整数列表就地排序", id(nlist_copy))
print(nlist_copy)
````

# 升序和降序
- ```sorted```和```list.sort```都有一个选择升序或降序的参数```reverse```
- 升序（缺省）```reverse=False```
- 降序```reverse=True```

````python
fsorted = sorted(flist, reverse=False)  # 升序排序
fsorted_rev = sorted(flist, reverse=True)  # 降序排序

plt.figure(figsize=(8,3), dpi=150)
plt.subplot(1, 2, 1)
plt.stem(range(total), fsorted)
plt.subplot(1, 2, 2)
plt.stem(range(total), fsorted_rev)
plt.show()
````

# ```sorted```和```list.sort```的对比
## 共同点
- 采用相同的排序算法，排序性能一样；
- 都可以对各种数据类型进行升序或降序的排序；

## 不同点
- ```sorted```返回一个新的列表对象；
- ```list.sort```在原列表就地排序，不产生新列表对象；

```python
nsorted = sorted(nlist)  # 返回排好序的列表，原列表保持不变
nlist.sort()  # 在原列表就地排序，数据项次序被改变
```  

- ```sorted```可以对各种容器类型（列表/元组/集合/字典）包含的数据项进行排序；
- ```list.sort```只是列表类型的方法；

```python
ntuple = (23, 34, 1, 100, 10, 7)  # 一个整数元组
nsorted = sorted(ntuple)  # 得到一个排好序的整数列表
sset = {"Alice", "Bob", "Smitch", "Lucy", "Carol"}  # 一个字符串集合
ssorted = sorted(sset)  # 得到一个排好序的字符串列表
```

- 如果不需要改变原列表，或者原列表不再需要了，用```sorted```能获得**更好的性能**。

# 小结
- 可以用```random```模块生成随机数据；
- 内置函数```sorted```可以对各种数据类型排序
- 列表方法```list.sort```可以进行列表中的就地排序
- 参数```reverse```可以控制升序或降序
- ```sorted```和```list.sort```可以用于不同场合
